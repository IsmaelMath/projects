 ### - Forma Padr√£o

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(6)
x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)
```

```python
plt.figure(figsize=(16, 7))# Tamanho da figura como um todo
plt.suptitle("Figura", fontsize=15)

plt.subplot(1, 3, 1)# Posi√ß√£o da figura funcionado estilo matrix
plt.plot(x, y1, color="k")# Eixos
plt.title("Subplot1", pad=10)
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.subplot(1, 3, 2)# Posi√ß√£o da figura funcionado estilo matrix
plt.plot(x, y2, color="r")# Eixos
plt.title("Subplot2", pad=10)
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.subplot(1, 3, 3)# Posi√ß√£o da figura funcionado estilo matrix
plt.plot(x, y3, color="g")# Eixos
plt.title("Subplot3", pad=10)
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.tight_layout(pad=4)# Espa√ßamento entre os plots
plt.show()
```

### - Forma simplificada

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 7))
fig.suptitle("Figure")
ax[0].plot(x, y1, color="black")
ax[1].plot(x, y2, color="red")
ax[2].plot(x, y3, color="green")

for i in range(3):
    ax[i].set(title = f"subplot {i+1}", xlabel="Eixo x", ylabel="Eixo y")
```
